```python
import pandas as pd 
import folium
from folium.plugins import HeatMap
```

```python
choques = pd.read_csv("Crash_Hotspots_MTY.csv")
choques.head()
```

```python
choques['LAT'] = choques["LAT"].astype(float)
choques['LONG'] = choques["LONG"].astype(float)
choques['cant_CP'] = choques["cant_CP"].astype(float)
choques=choques.dropna(axis=0, subset=['LAT', 'LONG', 'cant_CP'])
```

```python
max_amount = float(choques['cant_CP'].max())
```

```python
hmap = folium.Map(location=[25.70441, -100.28174], zoom_start=10, )
```

```python
HeatMap(choques, radius=10).add_to(hmap)
```
